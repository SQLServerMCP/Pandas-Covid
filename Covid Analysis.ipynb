```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("WHO-COVID-19-global-data.csv")
data["year"] = data["Date_reported"].str[0:4]
data["monthyear"] = data["Date_reported"].str[0:7]
```

```python
data.head()
```

```python
country = "The United Kingdom"
ukdata = data[data.Country == country]
#ukdata = ukdata[ukdata.year == "2021"]
```

```python
ukdata.head()
```

```python
x = ukdata["Date_reported"]
y = ukdata["New_cases"]
z = ukdata["monthyear"]

charttitle = f"Covid Cases in {country}"
colors = ['Gray'] 

plt.figure(figsize=(20,5))
plt.grid(True)
plt.bar(z, y, color = colors, width = 0.5)
plt.xlabel("Time", fontsize = 15)
plt.ylabel("Covid Cases", fontsize  = 15)
plt.title(charttitle, fontsize  = 20)
plt.xticks(rotation = 90,  fontsize = 15)
plt.yticks(rotation = 0,  fontsize = 15)
plt.show()
```
